## Municípios Nomes

Lê o arquivo csv com o nome dos municípios

```python
# Dataframe
df_mun = pd.read_csv(
    'https://raw.githubusercontent.com/michelmetran/sp/main/data/tabs/tab_municipio_nome.csv',
    usecols=['id_municipio', 'municipio_nome']
)
df_mun.head(3)
```

<br>

## Municípios IDs

```python
# Dataframe
df = pd.read_csv(
    os.path.join('data', 'tabs', 'input', 'tab_municipio_ids.csv'),
    usecols=['id_municipio', 'id_gaema', 'id_redeprotetiva'],
)
display(df.head(3))

# Merge Municípios
df = pd.merge(
    df_mun,
    df,    
    how='left',
    left_on='id_municipio',
    right_on='id_municipio',
)

# Results
display(df.head(3))
```

<br>

## Município GAEMAs

```python
# Dataframe GAEMAs
df_gaema = pd.read_csv(os.path.join('data', 'tabs', 'input', 'tab_gaema.csv'))
display(df_gaema.head(3))

# Merge GAEMAs
df = pd.merge(
    df,
    df_gaema,
    how='left',
    left_on='id_gaema',
    right_on='id_gaema',
)

# Results
display(df.head(3))
```

<br>

## Município Rede Protetiva

```python
# Dataframe Rede Protetiva
df_redeprotetiva = pd.read_csv(os.path.join('data', 'tabs', 'input', 'tab_redeprotetiva.csv'))
display(df_redeprotetiva.head(3))

# Merge Rede Protetiva
df = pd.merge(
    df,
    df_redeprotetiva,
    how='left',
    left_on='id_redeprotetiva',
    right_on='id_redeprotetiva',
)

# Results
display(df.head(3))
```

```python
# Results
df.to_csv(
    os.path.join('data', 'tabs', 'output', 'tab_municipio_gaema_rp.csv'),
    index=False,
)
df.head()
```
